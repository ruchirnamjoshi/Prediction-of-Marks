```python
# importing required libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# Reading data from remote link
url = "http://bit.ly/w-data"
df = pd.read_csv(url)

df.head(10)
```

```python
df.shape
```

```python
# Ploting the distribution of score
df.plot(x="Hours",y="Scores",style='p')
plt.title("Hours vs Percentage")
plt.xlabel("Hours Studied")
plt.ylabel("Percentage Scored")
plt.show()
```

# Preparing the data

```python
X = df.iloc[:,:-1].values
Y = df.iloc[:,1].values
```

Splitting data into training and testing dat

```python
from sklearn.model_selection import train_test_split
X_train , X_test , y_train, y_test = train_test_split(X,Y,test_size = 0.2,random_state=0)
```

# Training the Algorithm

```python
from sklearn.linear_model import LinearRegression
algo1=LinearRegression()
algo1.fit(X_train,y_train)

print("Training Complete!")
```

```python
# Plotting the regression line
line = algo1.coef_*X+algo1.intercept_

plt.scatter(X,Y)
plt.plot(X,line)
plt.show()
```

# Making Predictions

```python
print(X_test)
```

```python

y_predict = algo1.predict(X_test)
tab = pd.DataFrame({'Actual': y_test , 'Predicted': y_predict})
tab
```

```python
score_pred = np.array([9.25])
score_pred = score_pred.reshape(-1,1)
predict = algo1.predict(score_pred)
print("No of Hours={}".format(9.25))
print("Predicted Score={}".format(predict[0]))
```

# Evaluating The Model

```python
from sklearn import metrics
print("Mean absolute Error: ", metrics.mean_absolute_error(y_test,y_predict))
```
